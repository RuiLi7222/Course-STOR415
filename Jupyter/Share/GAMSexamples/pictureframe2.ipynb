```python
%%gams 

#define a set called I with two elements '1' and '2'
set I /1, 2/;

#define two scalars called NumLabor and NumMetal
scalar NumLabor /4000/, NumMetal /5000/;

#define parameters called price, laborUse and metalUse
parameters
       price(I) /1 2.25, 2 2.60/,   
       laborUse(I) /1 2, 2 1/,  
    metalUse(I) /1 1, 2 2/;

variable profit;

positive variables x(I);
#This defines two variables x('1') and x('2').


equations obj, labor, metal;

obj..
#profit =e= price('1')*x('1')+price('2')*x('2');
#profit =e= 2.25*x('1')+2.60*x('2');
profit =e=   sum(I, price(I)*x(I));


labor..
#2*x('1')+ 1*x('2') =l= 4000;
#laborUse('1')*x('1')+laborUse('2')*x('2') =l= NumLabor;
sum(I, laborUse(I)*x(I)) =l= NumLabor;

metal..
sum(I, metalUse(I)*x(I)) =l= NumMetal;

model picframe /all/;

#GAMS option limrow specifies the number 
#of equations to include in GAMS listing file
option limrow = 4;

#GAMS option solprint on/off: whether to 
#not to include the solution report part in the listing file
option solprint = on;

solve picframe using lp maximizing profit;

```
