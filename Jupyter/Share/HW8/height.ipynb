```python
%%gams

set i /1*7/;
set j /1*7/;

parameters p(i)/1 4, 2 8, 3 22, 4 25, 5 21, 6 14, 7 6/;

parameters q(j)/1 4, 2 9, 3 19, 4 26, 5 20, 6 16, 7 6/;

parameters c(i,j);

c(i,j)=abs(ord(i)-ord(j));

variables pi(i,j), z;

positive variables pi;

equation obj, balancei(i), balancej(j);

obj.. z =e= sum((i,j), pi(i,j)*c(i,j));

balancei(i).. sum(j,pi(i,j)) =e= p(i);

balancej(j).. sum(i,pi(i,j)) =e= q(j);

model heightdist /all/;

solve heightdist using lp minimizing z;
```
